# Learning PyMC with linear regression example

The aim here is to run through some of the core concepts in Bayesian modelling, seeing how we can implement them with PyMC. We will look at:
* **generating data** by sampling from the model where some parameters are specified (i.e. sampling from a conditional probabilty distribution). This could be useful if you are making predictions of a model, or generating data which can then be used for parameter recovery.
* **parameter estimation** Given a small amount of the data we just generated, we will conduct parameter recovery to see how well we can estimate the parameter values.
* **posterior predictive checks**, to examine how well a model fits the data. In this case it should fit the data well because it actually generated the data int he first place

### Do some intial set up

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pymc
print(pymc.__version__)
```

## PART 1: Data generation

### The generative probabilty model $P(x,y,\mu,y,m,c,\tau)$

For this example we assume values of $x$ are uniformly distributed between 0-10

$x \sim Uniform(0,10)$

The true value of $y$ is given by the standard linear regression equation $y=mx+c$

$\mu \leftarrow m.x + c$

But observations are noisy with precision $\tau$, which is the inverse variance $\tau = 1/\sigma^2$

$y \sim Normal(\mu, \tau)$

And we have priors over the parameters: slope, intercept, and tau

$m \sim Normal(0,1/100)$

$c \sim Normal(0,1/100)$

$\tau \sim Gamma(0.01, 0.01)$

### Define known parameters

```python
true_m = 1.0
true_c = 0.0
true_tau = 1.0/(1.5*1.5)
```

### Define the model in PyMC for data generation, $P(x,y|m,c,\tau)$

The following code creates PyMC objects, which embody our probabilistic model above

```python
def model2():
    # priors
    x = pymc.Uniform("x", lower=0, upper=10)
    tau = pymc.Gamma("tau", alpha=0.1, beta=0.1, value = true_tau, observed=True)

    m = pymc.Normal("m", mu=0.0, tau=1.0/100, value=true_m, observed=True)
    c = pymc.Normal("c", mu=0.0, tau=1.0/100, value=true_c, observed=True)

    @pymc.deterministic(plot=False)
    def mu(m=m, x=x, c=c):
        return m*x + c
    
    y = pymc.Normal("y", mu=mu, tau=tau)
    
    return locals()
```

```python
# generate MCMC object
m2 = pymc.MCMC(model2())
```

### Generate data with MCMC sampling

```python
m2.sample(iter=1000, burn=100)
```

```python
tempy = m2.y.trace()
tempx = m2.x.trace()
plt.plot(tempx,tempy,'o')
```

So now we have generated `(x,y)` from the model, basically by sampling many times from the distribution $P(x,y|m,c,\tau)$

## PART 2: parameter recovery using generated data. $P(m,c|x,y,tau)$

### Prep the generated data from

Our observed data in in `tempx` and `tempy`, but we generated many  samples. We want to see how good our inferences can be with just a handfull of x,y data points

```python
N=5
xdata = tempx[:N]
ydata = tempy[:N]
plt.plot(xdata,ydata,'o')
```

```python
print(xdata.size)
print(ydata.size)
```

### Define the model for sampling from $P(m,c|x,y,tau)$

```python
def model3():#(true_m, true_c, true_tau)
    # priors
    x = pymc.Uniform("x", lower=0, upper=10, value=xdata, observed=True)
    tau = pymc.Gamma("tau", alpha=0.1, beta=0.1, value = true_tau, observed=True)

    m = pymc.Normal("m", mu=0.0, tau=1.0/100)
    c = pymc.Normal("c", mu=0.0, tau=1.0/100)

    @pymc.deterministic(plot=False)
    def mu(m=m, x=x, c=c):
        return m*x + c
    
    y = pymc.Normal("y", mu=mu, tau=tau, value=ydata, observed=True)
    
    # posterior prediction. This will calculate values of x and y 
    # which are under the influence of the parameters, but are seperated 
    # from the data
    x_pred = pymc.Uniform("x_pred", lower=0, upper=10)
    y_pred = pymc.Normal("y_pred", mu=mu, tau=tau) # but the 'mu' IS tied to data 
    # if you really wanted though you could set up a set of uniformly spaced x vales
    # and calc mu for that, and get predited y values.  But the way how it is above 
    # will give a distribution of expected y values for each given provided x value.

    return locals()
```

```python
# generate MCMC object
m3 = pymc.MCMC(model3())
```

```python
m3.sample(iter=10000, burn=500)
```

### Examine the parameter estimates, and compare to the true parameters used

```python
pymc.Matplot.plot(m3.m)
```

```python
pymc.Matplot.plot(m3.c)
```

```python
plt.plot( m3.c.trace() , m3.m.trace() ,'.')
```

## PART 3: Use the fitted model to make predictions. Model checking.

The posterior predictive check was actually already generated in Part 2 where we did parameter estimation. But here we examine the predictions (in x/y data space).

```python
tempx = m3.x_pred.trace()
tempy = m3.y_pred.trace()
print(tempx.size)
print(tempy.size)
```
