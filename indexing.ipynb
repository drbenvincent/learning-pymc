Replicating the "Repeated measurement of IQ" example from [Lee & Wagenmakers (2014)](http://www.cambridge.org/gb/academic/subjects/psychology/psychology-research-methods-and-statistics/bayesian-cognitive-modeling-practical-course?format=PB), but with my own made up data.

As a beginer in both Python and PyMC, I'm using this example to understand how PyMC deals with variables that are 1D and 2D. Hopefully it's useful for other people as well.

```python
# do some set up
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pymc
```

## Define the data

```python
xdata = np.array([[92, 100, 113], 
              [99, 100, 120]])
I, J = xdata.shape
```

## Define the model

```python
def my_model():
    
    # NOTE TO SELF. It's important to declare these as objects
    mu = np.empty(I, dtype=object)
    x = np.empty([I,J], dtype=object)
    
    # priors
    tau = pymc.Gamma("tau", 0.01, 0.01)
    
    for i in range(I):
        mu[i] = pymc.Uniform('mu_%i' % i, lower=0, upper=300)
        
    # likelihood
    for i in range(I):
        for j in range(J):
            x[i,j] = pymc.Normal('x_%i,%i' % (i,j), mu=mu[i], tau=tau,
                                 value=xdata[i,j], observed=True)

    return locals()
```

## Make a PyMC MCMC object with the model, and generate samples

```python
# generate MCMC object
pymc_inference = pymc.MCMC(my_model())

# do the inference
burn=2000
pymc_inference.sample(iter=burn+5000, burn=burn)
```

## Visualise the chains of $\tau$ and $\mu$

```python
pymc.Matplot.plot( pymc_inference.tau )
```

```python
pymc.Matplot.plot( pymc_inference.mu[0] )
```

```python
pymc.Matplot.plot( pymc_inference.mu[1] )
```

## Extracting chain values for doing interesting things with

```python
mu0 = pymc_inference.trace('mu_0')[:]
print(mu0)
```

```python
mu1 = pymc_inference.trace('mu_1')[:]
print(mu1)
```

```python
tau = pymc_inference.trace('tau')[:]
print(tau)
```
